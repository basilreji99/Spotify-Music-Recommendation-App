```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
data = pd.read_csv('C:/Users/s.hussain/Downloads/spotify-2023.csv', encoding='latin1')

# Selecting the top 100 rows
smallerdata = data.head(500)

# Extracting the variables for the scatter plot
bpm= smallerdata['bpm']
danceability = smallerdata["danceability_%"]

# Creating the scatter plot
plt.figure(figsize=(10, 10))
plt.scatter(bpm, danceability, color='blue', alpha=0.5)
plt.title('beats per minute & danceability')
plt.ylabel('bpm')
plt.xlabel('danceability')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Provided dataset
# Read the dataset
data = pd.read_csv('C:/Users/s.hussain/Downloads/spotify-2023.csv', encoding='latin1')

# Selecting the top 100 rows
smallerdata = data.head(500)

# Creating histograms
plt.figure(figsize=(10, 6))

# Histogram for 'acousticness'
plt.subplot(1, 2, 1)
plt.hist(smallerdata['acousticness_%'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of acousticness')
plt.xlabel('acousticness_%')
plt.ylabel('Frequency')

# Histogram for 'released_month'
plt.subplot(1, 2, 2)
plt.hist(smallerdata['released_month'], bins=range(1, 14), color='lightgreen', edgecolor='black', align='left')
plt.title('Histogram of Released Month')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.xticks(range(1, 13))  # Set x-axis ticks to months (1 to 12)

plt.tight_layout()
plt.show()
```
